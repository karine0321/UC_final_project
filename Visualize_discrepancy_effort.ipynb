```python
# load state boundary
import geopandas as gpd
state_path = r'.\data\cb_2018_us_state_500k\cb_2018_us_state_500k.shp'
shape_state = gpd.read_file(state_path)
shape_state = shape_state[shape_state.NAME != 'Alaska']
shape_state = shape_state[shape_state.NAME != 'Hawaii']
shape_state = shape_state[shape_state.NAME != 'Commonwealth of the Northern Mariana Islands']
shape_state = shape_state[shape_state.NAME != 'United States Virgin Islands']
shape_state = shape_state[shape_state.NAME != 'Puerto Rico']
shape_state = shape_state[shape_state.NAME != 'Guam']
shape_state = shape_state[shape_state.NAME != 'American Samoa']

shape_state
```

```python
import geopandas as gpd
import os
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt


outputpath = "./data/discr_eff/"
state_discr_eff = pd.read_csv(os.path.join(outputpath, "discr_eff_val.csv"))
# state_discr_eff

integrate = pd.merge(shape_state[shape_state.NAME != 'District of Columbia'], state_discr_eff[list(state_discr_eff.columns)[1:]], left_on=["NAME"], right_on=["STATE"],how="left")
# integrate

integrate['judge_bed_covid'] = integrate.apply(lambda x: np.sign(x.bed_covid_discrepancy), axis = 1)
integrate['judge_pop_covid'] = integrate.apply(lambda x: np.sign(x.pop_covid_discrepancy), axis = 1)
integrate['judge_bed_pop'] = integrate.apply(lambda x: np.sign(x.bed_pop_discrepancy), axis = 1)
# integrate

base = integrate.plot(column='judge_bed_covid',cmap='viridis', scheme='User_Defined',classification_kwds=dict(bins=[-1,0,1]), figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/judge_bed_covid.png")

base = integrate.plot(column='judge_pop_covid',cmap='viridis', scheme='User_Defined',classification_kwds=dict(bins=[-1,0,1]),figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/judge_pop_covid.png")

base = integrate.plot(column='judge_bed_pop',cmap='viridis', scheme='User_Defined',classification_kwds=dict(bins=[-1,0,1]), figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/judge_bed_pop.png")

base = integrate.plot(column='bed_covid_discrepancy',cmap='Blues', scheme='quantiles', figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/bed_covid_discrepancy.png")

base = integrate.plot(column='pop_covid_discrepancy',cmap='Blues', scheme='quantiles', figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/pop_covid_discrepancy.png")

base = integrate.plot(column='bed_pop_discrepancy',cmap='Blues', scheme='quantiles', figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/bed_pop_discrepancy.png")

base = integrate.plot(column='bed_covid_effort',cmap='BuPu', scheme='quantiles', figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/bed_covid_effort.png")

base = integrate.plot(column='pop_covid_effort',cmap='BuPu', scheme='quantiles', figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/pop_covid_effort.png")

base = integrate.plot(column='bed_pop_effort',cmap='BuPu', scheme='quantiles', figsize=(22,10), legend=True, legend_kwds=dict(loc='lower left',  frameon=True))
shape_state.plot(ax=base, color='none', edgecolor='grey')
plt.savefig("./pic/bed_pop_effort.png")
```
